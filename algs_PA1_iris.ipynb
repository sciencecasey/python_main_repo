**Change the Species to Categorical variable for plotting**

```python
iris.dtypes
```

```python
iris.species.astype('category')
```

```python
iris.species = iris.species.astype('category')
```

```python
colors = iris.species.cat.codes
colors
```

```python
colors[colors==0] = 'red'
colors[colors==1] = 'purple'
colors[colors==2] = 'green'
```

```python
iris.plot.scatter('sepal_length', 'sepal_width', c=colors)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
s = matplotlib.patches.Patch(color = 'red', label='Setosa')
v = matplotlib.patches.Patch(color = 'blue', label = 'Versicolor')
vi = matplotlib.patches.Patch(color = 'green', label = 'Virginica')
plt.legend(handles = [s,v,vi])
```
